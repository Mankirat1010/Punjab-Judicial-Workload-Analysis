```python
import pandas as pd
```

```python
df_2010 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2010.csv")

df_2011 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2011.csv")

df_2012 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2012.csv")

df_2013 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2013.csv")

df_2014 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2014.csv")

df_2015 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2015.csv")

df_2016 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2016.csv")

df_2017 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2017.csv")

df_2018 = pd.read_csv("C:/Users/HP/OneDrive/Desktop/Projects/Judicial_DA/data/justice_data/csv/cases/cases_2018.csv")
```

# state code of punjab is 22 (checked from keys..)

```python
df_punjab_2010 = df_2010[df_2010['state_code'] == 22]

df_punjab_2011 = df_2011[df_2011['state_code'] == 22]

df_punjab_2012 = df_2012[df_2012['state_code'] == 22]

df_punjab_2013 = df_2013[df_2013['state_code'] == 22]

df_punjab_2014 = df_2014[df_2014['state_code'] == 22]

df_punjab_2015 = df_2015[df_2015['state_code'] == 22]

df_punjab_2016 = df_2016[df_2016['state_code'] == 22]

df_punjab_2017 = df_2017[df_2017['state_code'] == 22]

df_punjab_2018 = df_2018[df_2018['state_code'] == 22]
```

# total number of cases in punjab yearwise

```python
print('2010 ->', len(df_punjab_2010))
print('2011 ->', len(df_punjab_2011))
print('2012 ->', len(df_punjab_2012))
print('2013 ->', len(df_punjab_2013))
print('2014 ->', len(df_punjab_2014))
print('2015 ->', len(df_punjab_2015))
print('2016 ->', len(df_punjab_2016))
print('2017 ->', len(df_punjab_2017))
print('2018 ->', len(df_punjab_2018))
```

# merging dataframes of all cases from 2010 to 2018 into one

```python
df_punjab = pd.concat([
    df_punjab_2010, 
    df_punjab_2011, 
    df_punjab_2012, 
    df_punjab_2013, 
    df_punjab_2014, 
    df_punjab_2015, 
    df_punjab_2016, 
    df_punjab_2017, 
    df_punjab_2018], ignore_index=True)
```
