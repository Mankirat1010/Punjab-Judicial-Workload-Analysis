### Converting dates dtype from STRING to DATE

# Removing Outliers

### 1). date_of_filing > date_of_decision

```python
len(punjab_cases[punjab_cases['date_of_filing'] > punjab_cases['date_of_decision']])
```

```python
punjab_cases = punjab_cases[
    (punjab_cases['date_of_decision'].isna()) |
    (punjab_cases['date_of_filing'] <= punjab_cases['date_of_decision'])
]
```
