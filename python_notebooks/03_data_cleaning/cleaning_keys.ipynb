# Formatting columns where needed

### 1. Converting judge IDs from float to int (judge_case_merge_key)

```python
judge_case_merge_key['ddl_filing_judge_id'] = pd.to_numeric(
    judge_case_merge_key['ddl_filing_judge_id'], errors='coerce'
).astype('Int64')

judge_case_merge_key['ddl_decision_judge_id'] = pd.to_numeric(
    judge_case_merge_key['ddl_decision_judge_id'], errors='coerce'
).astype('Int64')
```

```python
judge_case_merge_key
```

### 2. Formatting dates in judges_clean 

```python
# Convert dates to datetime
judges_clean['start_date'] = pd.to_datetime(
    judges_clean['start_date'],
    format='%d-%m-%Y',
    errors='coerce'
)

judges_clean['end_date'] = pd.to_datetime(
    judges_clean['end_date'],
    format='%d-%m-%Y',
    errors='coerce'
)
```

```python
judges_clean['start_date']
```

### 3. Code columns in key tables - STR to INT

```python
cases_state_key['state_code'] = pd.to_numeric(cases_state_key['state_code'], errors='coerce').astype('Int64')

cases_district_key['state_code'] = pd.to_numeric(cases_district_key['state_code'], errors='coerce').astype('Int64')
cases_district_key['dist_code'] = pd.to_numeric(cases_district_key['dist_code'], errors='coerce').astype('Int64')

cases_court_key['state_code'] = pd.to_numeric(cases_court_key['state_code'], errors='coerce').astype('Int64')
cases_court_key['dist_code'] = pd.to_numeric(cases_court_key['dist_code'], errors='coerce').astype('Int64')
cases_court_key['court_no'] = pd.to_numeric(cases_court_key['court_no'], errors='coerce').astype('Int64')

judges_clean['state_code'] = pd.to_numeric(judges_clean['state_code'], errors='coerce').astype('Int64')
judges_clean['dist_code'] = pd.to_numeric(judges_clean['dist_code'], errors='coerce').astype('Int64')
judges_clean['court_no'] = pd.to_numeric(judges_clean['court_no'], errors='coerce').astype('Int64')
```
